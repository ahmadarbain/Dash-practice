```python
from jupyter_dash import JupyterDash
from dash.dependencies import Output, Input
import dash_core_components as dcc
import dash_html_components as html
import dash_bootstrap_components as dbc
import pandas as pd
```

```python
proverty_data = pd.read_csv("/PROJECT/DASH_PLOTLY/Interactive-Dashboard-Dash-Plotly/povstats-csv-zip-242-kb-/PovStatsData.csv")
proverty_data.head()
```

```python
app = JupyterDash(__name__)
```

```python
app.layout = html.Div([
    dcc.Dropdown(id = "country",
                 options = [{"label":country, "value":country}
                           for country in proverty_data["Country Name"].unique()]),
    
    html.Div(id="report")
])
```

```python
@app.callback(Output("report", "children"),
              Input("country", "value"))

def display_selected_country(country):
    if country is None:
        return ""
    
    filtered_df = proverty_data[(proverty_data["Country Name"]==country)&(proverty_data["Indicator Name"]=="Population, total")]
    population = filtered_df.loc[:, '2010'].values[0]
    return [html.H3(country),
            f'The population of {country} in 2010 was {population:,.0f}.']
```

```python
if __name__ == "__main__":
    app.run_server(mode="inline", width="80%")
```
